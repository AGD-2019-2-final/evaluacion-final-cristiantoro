```python
!rm -rf input output
!mkdir output
%load_ext bigdata
%pig_start
%timeout 300
!hadoop fs -rm data.tsv
!hadoop fs -put data.tsv
!hadoop fs -ls output/*
!hadoop fs -rm output/*
!hadoop fs -rmdir output
```

```python
%%pig
data = LOAD 'data.tsv' 
   AS (f1:CHARARRAY, f2:BAG{t:tuple(p:CHARARRAY)},f3:MAP[]);
DUMP data;
```

```python
%%pig
r = FOREACH data GENERATE f1,COUNT(f2.p) AS f4, SIZE(f3) AS f5;
DUMP r;
```

```python
%%pig
top= ORDER r BY f1,f4,f5;
DUMP top;
```

```python
%%pig
STORE top INTO 'output' USING PigStorage(',');
fs -get output/ .
```
