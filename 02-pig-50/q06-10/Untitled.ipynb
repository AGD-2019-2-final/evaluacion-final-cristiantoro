```python
!rm -rf input output
!mkdir output
%load_ext bigdata
%pig_start
%timeout 300
!hadoop fs -rm data.tsv
!hadoop fs -put data.tsv
!hadoop fs -ls output/*
!hadoop fs -rm output/*
!hadoop fs -rmdir output
```

```python
%%pig
data = LOAD 'data.tsv' 
   AS (f1:CHARARRAY, f2:BAG{t:(p:CHARARRAY)},f3:MAP[]);
DUMP data;
```

```python
%%pig
r = FOREACH data GENERATE FLATTEN(f3) AS f4;
DUMP r;
```

```python
%%pig
grupo = GROUP r by $0;
DUMP grupo;
```

```python
%%pig
conteo = FOREACH grupo GENERATE group, COUNT($1);
DUMP conteo;
```

```python
%%pig
STORE conteo INTO 'output' USING PigStorage(',');
fs -get output/ .
```
