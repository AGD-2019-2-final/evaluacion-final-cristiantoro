```python
!rm -rf input output
!mkdir output
%load_ext bigdata
%pig_start
%timeout 300
!hadoop fs -rm data.tsv
!hadoop fs -put data.tsv
!hadoop fs -ls output/*
!hadoop fs -rm output/*
!hadoop fs -rmdir output
```

```python
%%pig
data = LOAD 'data.tsv' 
   AS (f1:CHARARRAY, f2:CHARARRAY, f3:INT);
DUMP data;
```

```python
%%pig
numero= ORDER data BY f3 ASC;
DUMP numero
```

```python
%%pig
top = FOREACH numero GENERATE f3;
DUMP top;
```

```python
%%pig
top2 = LIMIT top 5;
DUMP top2;
```

```python
%%pig
STORE top2 INTO 'output';
fs -get output/ .
```
