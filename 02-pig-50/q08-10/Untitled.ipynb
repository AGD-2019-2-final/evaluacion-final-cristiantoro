```python
!rm -rf input output
!mkdir output
%load_ext bigdata
%pig_start
%timeout 300
!hadoop fs -rm data.tsv
!hadoop fs -put data.tsv
!hadoop fs -ls output/*
!hadoop fs -rm output/*
!hadoop fs -rmdir output
```

```python
%%pig
data = LOAD 'data.tsv' 
   AS (f1:CHARARRAY, f2:BAG{t:tuple(p:CHARARRAY)},f3:MAP[]);
DUMP data;
```

```python
%%pig
r = FOREACH data GENERATE FLATTEN(f2) AS f5,FLATTEN(f3) AS f4;
DUMP r;
```

```python
%%pig
grupo = GROUP r BY (f5,f4);
DUMP grupo;
```

```python
%%pig
top = FOREACH grupo GENERATE group, COUNT($1);
DUMP top;
```

```python
%%pig
STORE top INTO 'output';
fs -get output/ .
```
